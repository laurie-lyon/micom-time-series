## Subdoligranulum promoting Faecalibacterium - metabolite exchange

I have detected a significant positive interaction between genera Subdoligranulum and Faecalibacterium in M01 in multiple iterations of the TLC lagged regression analysis, so I want to make a figure from the micom grow output (AGORA2, wd, tradeoff=0.8) showing which metabolites have the highest flux from OTUj -> OTUi

```python
import pandas as pd
from micom.workflows.results import load_results

results = load_results("../data/growth_rates/growth_M01_agora201_gurobi_wd_08_2.zip")
print(results.growth_rates.head())
```

## Compute focal interactions for Subdoligranulum

```python
from micom.interaction import interactions


ints = interactions(results, taxa="g__Subdoligranulum")
ints.sort_values(by="flux", ascending=False).head()
```

```python
ints_sub_vs_faec = ints[ints["partner"].str.contains("g__Faecalibacterium")]
ints_sub_vs_faec.sort_values(by="flux", ascending=False).head()
ints_sub_vs_faec.to_csv("../data/interactions/ints_subdoligranulum_faecalibacterium.csv", index=False)
```

```python
from micom.interaction import summarize_interactions

ints_sub_vs_faec_summary = summarize_interactions(ints_sub_vs_faec)
ints_sub_vs_faec_summary.head()

#export to csv, making sure to include metabolite names that are in the index column
#ints_sub_vs_faec_summary.to_csv("../data/interactions/ints_subdoligranulum_faecalibacterium.csv", index=False)


```
