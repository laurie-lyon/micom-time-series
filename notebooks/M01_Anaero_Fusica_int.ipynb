I have detected a significant negative interactions between genera Anaerostipes and Fusicatenibacter in M01 in multiple iterations of the TLC lagged regression analysis, so I want to make a figure from the micom grow output (AGORA2, wd, tradeoff=0.8) showing which metabolites have the highest flux from OTUj -> OTUi

```python
import pandas as pd
from micom.workflows.results import load_results

results = load_results("../data/growth_rates/growth_M01_agora201_gurobi_wd_08_2.zip")
print(results.growth_rates.head())
```
